```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences

sent = [
    'I believe in constant growth',
    'I believe in constant Decline ',
    'Practice makes perfect',
    'Coding makes ideas come alive',
    'Success comes from daily effort',
    'Stay curious and keep exploring',
    'Hard work always beats talent',
    'Believe in yourself',
    'Focus brings clarity and direction',
    'Stay focused always',
]
```

```python
sent
```

```python
# define the vacalbulary size 
voc_size = 10000
```

```python
# One Hot Representation 
one_hot_representation=[one_hot(words,voc_size)for words in sent]
one_hot_representation

# o/p = only index
```

```python
## Word Embedding Representation

from tensorflow.keras.layers import Embedding
# from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 10
embedded_doc = pad_sequences(one_hot_representation,padding='pre', maxlen=sent_length)
print(embedded_doc)
```

```python
# feature representation
dim =10
```

```python
model=Sequential()
model.add(Embedding(voc_size, dim,input_length=sent_length))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
# o/p = 10 dimention 
```
